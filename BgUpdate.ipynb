```python
from PIL import Image
from rembg import remove

# Load your original image
input_path = "input_image.jpg"
original_image = Image.open(input_path)

# Remove the background
image_no_bg = remove(original_image)

# Open a new background image and resize it to match the original image's dimensions
background_path = "bg_image.jpg"
new_background = Image.open(background_path).resize(original_image.size)

# Paste the image without background onto the new background
new_background.paste(image_no_bg, (0, 0), image_no_bg)

# Save the final image
new_background.save("output_image.jpg")
```
